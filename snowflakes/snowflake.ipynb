```python
from random import randint, choice, random


mode = 1 # 1 - large, 2 - composite
W,H = 1000,1000
ANIMATE = False

NEXTFRAME = 60 # number of frames till redraw

def setup():
    createCanvas(1000, 1000)
    
    # GUI element

    
    
    # for timing animations
    global nextFrame
    nextFrame = NEXTFRAME

    
def draw():

    global nextFrame
    
    nextFrame-=1
    if nextFrame==0 or not ANIMATE:
        nextFrame = NEXTFRAME

        if mode == 1:
            
                background(158, 149, 222)
                
                #rect(100,100,100,10)
                
                line(500,0,500,1000)
                line(0,500,1000,500)
                arrow_flake()   
                
                                                    
        elif mode == 2:
            background(158, 149, 222)
            translate(75,75)
            
            for y in range(0,500,70):
                for x in range(0,500,70):
                    push()
                    translate(x*1.5,y*1.5)
                    scale(0.1)
                    arrow_flake()
                    pop()
                    
        if not ANIMATE:
            noLoop()
            

      
```

```python
def arrow_flake_conf():
    # some styles
    conf = {}
    
    conf['end_cap'] = choice([PROJECT,ROUND])
    conf['wide_stroke'] = map(random()*20,0,20,10,20)
    conf['light_stroke'] = map(random()*8,0,8,7,8)
        
    # number of branches
    conf['branches'] = choice([6,8])
    
    # lengths of stuff
    conf['y_h'] = randint(150,450)
    conf['leafs'] = randint(2,5)
    conf['leaf_len'] = randint(80,160)
    conf['leaf_alpha'] = randint(30,50)
    
    return conf
```

```python
      
                        
def display(x,y,info):
    push()
    translate(x,y)
    
    dy = 20
    y = 0
    
    for item in info:
        y += dy
        text_str = "%s %s" % (item[0],item[1])
        text(text_str, 0,y)                      
    pop()                                    

conf = arrow_flake_conf()

def arrow_flake():
    global branches, conf
    # assume size of 1000 and scale accordingly
    # start in middle
    translate(500,500)
    
    push()

    # some styles
    global conf
   
    #end_cap = choice([PROJECT,ROUND])
    #wide_stroke = map(random()*20,0,20,10,20)
    #light_stroke = map(random()*8,0,8,7,8)
        
    # number of branches
    #branches = choice([6,8])
    
    # lengths of stuff
    #y_h = randint(150,450)
    #leafs = randint(2,5)
    #leaf_len = randint(80,160)
    #leaf_alpha = randint(30,50)
    
    info = [('end_cap', conf['end_cap']),
            ('wide_stroke', conf['wide_stroke']),
            ('light_stroke', conf['light_stroke']),
            ('branches', conf['branches']),
            ('y_h', conf['y_h']),
            ('leafs', conf['leafs']),
            ('leaf_len', conf['leaf_len']),
            ('leaf_alpha', conf['leaf_alpha'])]
    
    
    
    # main line style
    stroke(randint(200,255),randint(200,255),randint(200,255))
    strokeCap(conf['end_cap'])
    
    # branches
    for _ in range(conf['branches']):
        strokeWeight(conf['wide_stroke'])

        # main branch
        line(0,0,0,-conf['y_h'])
        
        # leaves
        push()
        for _ in range(conf['leafs']):
            translate(0,-conf['y_h']/conf['leafs'])
            push()
            rotate(-conf['leaf_alpha'])
            strokeWeight(conf['light_stroke'])
            line(0,0,0,conf['leaf_len'])
                        
            pop()
            
            push()
            rotate(conf['leaf_alpha'])
            strokeWeight(conf['light_stroke'])
            line(0,0,0,conf['leaf_len'])
            pop()
        
        pop()
        
        #break
        rotate(radians(360/conf['branches']))
        
    rotate(radians(360/conf['branches']))
    pop()
    
    fill(0)
    #rect(-10,-10,20,20)
    
    # middle can be a shape or intersecting lines
    if 0 == 0:
        # little flower shape
        # was suppose to be a hexagon
        # 
        push()
        
        
        translate(-25,-25)
        beginShape();
        
        beginShape()
        fill(0, 0, 0)
        noStroke()
        vertex(-50, 0)
        for _ in range(conf['branches']):
            vertex(50, 0)
            line(-50,0,50,0)
            rotate(radians(360/conf['branches']))
        endShape(CLOSE)
                    
        pop()
        
    display(-400,-400,info)
    
    
```
